```python
#procedemos a hacer las "Trasformaciones' necesarias para poder usar los datasets y nos den la informacion mas 
#optima en nuestros postrerior analisis de los datos.
```

```python
#Generamos el cambio en id para cada plataforma agregando la primera letra del nombre de la plataforma
amazon_prime["show_id"] = amazon_prime["show_id"].apply(lambda x: "a" + x)
amazon_prime.head()
```

```python
disney_plus["show_id"] = disney_plus["show_id"].apply(lambda x: "d" + x)
disney_plus.head()
```

```python
hulu_titlles["show_id"] = hulu_titlles["show_id"].apply(lambda x: "h" + x)
hulu_titlles.head()
```

```python
netflix_titles["show_id"] = netflix_titles["show_id"].apply(lambda x: "n" + x)
netflix_titles.head()
```

```python
# Cambiamos los valores nulos por una 'G' en en la variable 'rating' todos los datasets
amazon_prime['rating'].fillna('G', inplace=True)
amazon_prime.head()
```

```python
disney_plus['rating'].fillna('G', inplace=True)
```

```python
hulu_titlles['rating'].fillna('G', inplace=True)
```

```python
netflix_titles['rating'].fillna('G', inplace=True)
```

```python
# hacemos una verificacion de que todos los nulos de la columna "rating" han sido cambiados.
amazon_prime['rating'].isna().count()
```

```python
#cambiamos el formato de la fecha AA-mm-dd en todos los Datasets
amazon_prime['date_added'] = pd.to_datetime(amazon_prime['date_added'], format='%B %d, %Y').dt.strftime('%Y-%m-%d')
```

```python
#Verificamos el cambio de fecha
amazon_prime.head()
```

```python
disney_plus['date_added'] = pd.to_datetime(disney_plus['date_added'], format='%B %d, %Y').dt.strftime('%Y-%m-%d')
```

```python
hulu_titlles['date_added'] = pd.to_datetime(hulu_titlles['date_added'], format='%B %d, %Y').dt.strftime('%Y-%m-%d')
```

```python
hulu_titlles.head()
```

```python
netflix_titles['date_added'] = netflix_titles['date_added'].str.strip()
netflix_titles['date_added'] = pd.to_datetime(netflix_titles['date_added'], format='%B %d, %Y').dt.strftime('%Y-%m-%d')
```

```python
#cambiamos todos los campos  a  letras "minusculas" en todos los datasets
amazon_prime = amazon_prime.applymap(lambda x: str(x).lower() if type(x) == str else x)
```

```python
#verificamos los cambios
amazon_prime.head()
```

```python
disney_plus = disney_plus.applymap(lambda x: str(x).lower() if type(x) == str else x)
```

```python
hulu_titlles = hulu_titlles.applymap(lambda x: str(x).lower() if type(x) == str else x)
```

```python
netflix_titles = netflix_titles.applymap(lambda x: str(x).lower() if type(x) == str else x)
```

```python
#Covertimos el campo "duration en dos campos para todos los datasets.
amazon_prime['duration'] = amazon_prime['duration'].str.replace('seasons', 'season')
```

```python
#Verificamos el cambio
amazon_prime.head()
```

```python
amazon_prime[['duration_int','duration_type']] = amazon_prime['duration'].str.split(n=1,expand=True)
amazon_prime['duration_int'] = amazon_prime['duration_int'].astype(int)
```

```python
disney_plus['duration'] = disney_plus['duration'].str.replace('seasons', 'season')
disney_plus[['duration_int','duration_type']] = disney_plus['duration'].str.split(n=1,expand=True)
disney_plus['duration_int'] = disney_plus['duration_int'].astype(int)
```

```python
hulu_titlles['duration'] = hulu_titlles['duration'].str.replace('seasons', 'season')
hulu_titlles['duration'].fillna("0 min", inplace=True)
```

```python
netflix_titles['duration'] = netflix_titles['duration'].str.replace('seasons', 'season')
netflix_titles['duration'].fillna("0 min", inplace=True)
netflix_titles[['duration_int','duration_type']] = netflix_titles['duration'].str.split(n=1,expand=True)
netflix_titles['duration_int'] = netflix_titles['duration_int'].astype(int)
```
